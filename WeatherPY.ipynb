```python
#Declare dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from api_keys import api_key

#Import Key
import api_keys

#Export the cities file as a CSV
output_data="cities.csv"

#Limit the ranges of latitudes and longitudes
longs_R=(-180,80)
lats_R=(-90,90)
```

```python
#create lists to save the latitudes,longitudes and cities and 
#create a command to determine the nearest cities

#To determine the city use "CITYPY"
from citipy import citipy

lats_longs=[]
cities=[]

#Generate a random latitude and longitude commbinations
lats=np.random.uniform(low=-90.000, high=90.000, size=1500)
longs=np.random.uniform(low=-180.000, high=180.000, size=1500)
lats_longs=zip(lats,longs)

for lats_long in lats_longs:
    city=citipy.nearest_city(lats_long[0],lats_long[1]).city_name

#to create the correct URLS we need to add "%20" instead of spaces
#in order to have the correct URLS
    city=city.replace(" ","%20")    
    
    if city not in cities:
        cities.append(city)
len(cities)

#We will add the city only if it does not exist already in the list of cities
```

```python
#Perform a weather check on each city using a series of successive API calls.
#Include a print log of each city as it'sbeing processed (with the city number and city name).
api_key = api_keys.api_key

url="http://openweathermap.org/data/2.5/weather?units=Imperial&appid=b6907d289e10d714a6e88b30761fae22"
#+ api_key
#b6907d289e10d714a6e88b30761fae22
```

```python
city_name=[]
cloudiness=[]
country=[]
date=[]
humidity=[]
lat=[]
long=[]
max_t=[]
wind_s=[]

record=1

print(f"Obtaining info from OpenWeatherMap API")
print(f"===================================")

# Loop through all cities in the list
for city in cities:  
#city   
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_t.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        long.append(response["coord"]["lon"])
        wind_s.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {record} | {city_record}")
        print(f"{url}&q={city}")
        
        # Add to counter
        record= record + 1

        
    # Skip if no match found
    except:
        print("No match found. Skipping to next city.")
    continue
```

```python
Weather_Dict={
    "City":city_name,
    "Cloudiness":cloudiness,
    "Country":country,
    "Date":date,
    "Humidity":humidity,
    "Lat":lat,
    "Long":long,
    "Max Temp":max_t,
    "Wind Speed":wind_s
}

WeatherData=pd.DataFrame(Weather_Dict)

WeatherData.count()
```

```python
#Put data into a CSV file
WeatherData.to_csv('WeatherData.csv')

#Check how DataFrame is shown
WeatherData.head()
```
